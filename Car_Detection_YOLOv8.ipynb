# 3. Training YOLO

Load configuaration files and pre-trained model

```python
from ultralytics import YOLO

# Load a model
yolo_yaml_path = 'yolo_data/data.yml'
model = YOLO('yolov8s.yaml').load('yolov8s.pt')
```

Set up important parameters

```python
epochs = 15 # S ln lp
imgsz = 540
batch_size = 8
patience = 5
lr = 0.0005
```

Start training (fine-tuning)

```python
results = model.train(
    data=yolo_yaml_path,
    epochs=epochs,
    imgsz=imgsz,
    batch=batch_size,
    lr0=lr,
    patience=patience,
    project='models',
    name='yolov8/detect/train'
)
```

# 4. Evaluate model

```python
from ultralytics import YOLO

model_path = 'models/yolov8/detect/train/weights/best.pt'
model = YOLO(model_path)

metrics = model.val(
    project='models',
    name='yolov8/detect/val'
)
```
